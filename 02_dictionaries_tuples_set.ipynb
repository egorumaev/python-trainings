```python
--5
import sys

lst_in = list(map(str.strip, sys.stdin.readlines()))

lst = [i.split() for i in lst_in]

d = {}

for number, key in lst:
    if key in d:
        d[key].append(number)
    else:
        d[key] = []
        d[key].append(number)

print(*sorted(d.items()))
```

```python
--9
import sys

lst_in = list(map(str.strip, sys.stdin.readlines()))

d = {}

for i in lst_in:
    key, value = i.split()
    if key not in d:
        d[key] = value
    else:
        d[key] += f', {value}'

[print(f'{key}: {value}') for key, value in d.items()]
```

```python
--10
n = int(input()) * 1000
things = {
    'карандаш': 20, 'зеркальце': 100, 'зонт': 500, 'рубашка': 300,
    'брюки': 1000, 'бумага': 200, 'молоток': 600, 'пила': 400,
    'удочка': 1200, 'расческа': 40, 'котелок': 820, 'палатка': 5240,
    'брезент': 2130, 'спички': 10
}

list_weight = list(things.values())
list_weight.sort(reverse = True)

weight_up_to_10 = 0
to_get_index = []

for i in list_weight:
    if weight_up_to_10 + i > n:
        continue
    else:
        weight_up_to_10 += i
        to_get_index.append(i)

res = [key for i in to_get_index for key, value in things.items() if i == value]

print(*res)
```
