```python
--10
seq = input().split()

latin = ['a', 'b', 'c', 'd', 'e', 'f',
        'g', 'h', 'i', 'j', 'k', 'l',
        'm', 'n', 'o', 'p', 'q', 'r',
        's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

def correct_email(email):
    if '@' not in email:
        return False
    flag_1, flag_2 = True, True
    part_1, part_2 = email.split('@')
    if (part_2.count('.') > 1 or '.' not in part_2) and '.' in part_2[0]:
        return False
    for i in part_1:
        if i not in latin and not i.isdigit() and i != '_':
            flag_1 = False
            break
    for i in part_2:
        if i not in latin and not i.isdigit() and i not in ['.', '_']:
            flag_1 = False
            break
    return flag_1 and flag_2

res = list(filter(correct_email, seq))

print(*res)
```

```python
--25
def get_add(a, b):
    if isinstance(a, str) and isinstance(b, str):
        return a + b
    elif type(a) in (int, float) and type(b) in (int, float):
        return a + b
    else:
        return None
```

```python
--26
def get_sum(it):
    res = 0
    if type(it) != dict:
        for i in it:
            if type(i) == int:
                res += i
        return res
    else:
        for i in range(len(it)):
            if type(it[i]) == int:
                res += it[i]
        return res
    return res
```

```python
--33
def is_free(lst):
    flag = False
    for i in lst:
        res = any(map(lambda x: x == '#', i))
        if res == True:
            flag = True
            break
    return flag
```
