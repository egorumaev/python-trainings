```python
--1
a, b = [float(input()) for i in range(2)]

def body_mass_index(mass, height):
    bm_index = mass / (height * height)
    if bm_index < 18.5:
        return 'Недостаточная масса'
    elif bm_index > 25:
        return 'Избыточная масса'
    else:
        return 'Оптимальная масса'

print(body_mass_index(a, b))
```

```python
--3
# вывести знак зодиака для введенного значения года
year = int(input())

zodiac_list = [
    'Дракон', 'Змея', 'Лошадь', 'Овца', 'Обезьяна', 'Петух',
    'Собака', 'Свинья', 'Крыса', 'Бык', 'Тигр', 'Заяц']

def zodiac_sign(year):
    if year >= 2000:
        zod_index = int((year - 2000) % 12)
        return zodiac_list[zod_index]
    else:
        zod_index = int(12 - abs(year - 2000) % 12)
        return zodiac_list[zod_index]

print(zodiac_sign(year))
```

```python
--4
# 1-й вариант решения

# изменитm порядок последних пяти цифр введеннего числа на обратный,
# удалив незначащие нули в начале измененного числа
number = input()

def number_reversal(number):
    if len(number) == 5:
        number = number[::-1]
        while number[0] == '0':
            number = number[1:]
        return int(number)
    else:
        res_1 = number[0]
        res_2 = number[1:][::-1]
        return int(res_1 + res_2)

print(number_reversal(number))
```

```python
--10
seq = [int(x) for x in input().split()]

def reverse_sequence(seq):
    if len(seq) == 1:
        return seq
    elif len(seq) % 2 == 0:
        for i in range(0, len(seq) - 1, 2):
            seq[i], seq[i + 1] = seq[i + 1], seq[i]
        return seq
    else:
        seq_end = [seq.pop(-1)]
        for i in range(0, len(seq) - 1, 2):
            seq[i], seq[i + 1] = seq[i + 1], seq[i]
        return seq + seq_end

print(*reverse_sequence(seq))
```

```python
--13
# 1-й вариант решения
def product_nums():
    n = int(input())
    num_set = [int(input()) for _ in range(n)]
    number = int(input())
    for i in range(len(num_set)):
        num_set_copy = num_set.copy()
        current_number = num_set_copy.pop(i)
        for j in num_set_copy:
            if current_number * j == number:
                return 'ДА'
    return 'НЕТ'

print(product_nums())
```

```python
--14
# 2-й вариант решения
def product_nums():
    n = int(input())
    num_set = [int(input()) for _ in range(n)]
    number = int(input())
    for i in range(len(num_set)):
        for j in range(i + 1, len(num_set)):
            if num_set[i] * num_set[j] == number:
                return 'ДА'
    return 'НЕТ'

print(product_nums())
```

```python
--18
# вывести номера умных холодильников, зараженных вирусом
frig = [
    'osfjwoiergwoignaewpjofwoeijfnwfonewfoignewtowenffnoeiwowjfninoiwfen',
    'anton',
    'aoooooooooontooooo',
    'elelelelelelelelelel',
    'ntoneeee',
    'tonee',
    '253235235a5323352n25235352t253523523235oo235523523523n',
    'antoooooooooooooooooooooooooooooooooooooooooooooooooooon',
    'unton']

res = []

for i in range(len(frig)):
    count = 0
    if 'a' in frig[i] and 'n' in frig[i] and 't' in frig[i] and 'o' in frig[i]:
        if 'a' in frig[i]:
            count += 1
            a_index = frig[i].find('a')
        if 'n' in frig[i][a_index + 1:]:
            count += 1
            n_index = frig[i].find('n', a_index + 1, len(frig[i]) - 1)
        if 't' in frig[i][n_index + 1:]:
            count += 1
            t_index = frig[i].find('t', n_index + 1, len(frig[i]) - 1)
        if 'o' in frig[i][t_index + 1:]:
            count += 1
            o_index = frig[i].find('o', t_index + 1, len(frig[i]) - 1)
        if 'n' in frig[i][o_index + 1:]:
            count += 1
        if count == 5:
            res.append(i + 1)

print(*res)
```

```python
--19
word = 'яндекс' + ' запретил букву '

# определим и отсортируем уникальные буквы
temp = word[:].replace(' ', '')

letters = []

for letter in temp:
    if letter not in letters:
        letters.append(letter)
letters.sort()

for letter in letters:
    print(word + letter)
    word = word.replace(letter, '')
    while word[0] == ' ' and letter != letters[-1]:
        word = word[1:]
    while '  ' in word:
        word = word.replace('  ', ' ')
```
