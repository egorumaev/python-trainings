```python
--4
a = int(input())

gen_abs = (abs(x) for x in range(-a, a + 1))

gen_cube = (x ** 3 for x in gen_abs)

for index, value in enumerate(gen_cube):
    print(value, end = ' ')
    if index == 3:
        break
```

```python
--5
from string import ascii_lowercase

lst = ascii_lowercase

gen = (i + j for i in lst for j in lst)

res = []

for index, value in enumerate(gen):
    res.append(value)
    if index == 49:
        break

print(*res)
```

```python
--7
a, b = map(int, input().split())

gen_nums = (x / 100. for x in range(a * 100, (b + 1) * 100))

gen = (0.5 * pow(x, 2) - 2.0 for x in gen_nums)

for i in range(20):
    print(round(next(gen), 2), end = ' ')
```

```python
--9
N = int(input())

def get_sum(N):
    summ = 0
    for i in range(1, N + 1):
        summ += i
        yield summ
```

```python
--10
import sys
import random
# установка "зерна" датчика случайных чисел, чтобы получались одни и те же случайные величины
random.seed(1)

# считывание списка из входного потока
lst_in = list(map(str.strip, sys.stdin.readlines()))

lst = [list(map(int, x.split())) for x in lst_in]

temp = list(zip(*lst))

random.shuffle(temp)

res = zip(*temp)

for i in res:
    print(*i)
```
