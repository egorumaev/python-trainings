```python
--1
cmd = input()

match cmd:
    case 'top' | 'Top' | 'TOP':
        print('Команда ' + cmd.lower())
    case 'bottom' | 'Bottom' | 'BOTTOM':
        print('Команда ' + cmd.lower())
    case 'right' | 'Right' | 'RIGHT':
        print('Команда ' + cmd.lower())
    case 'left' | 'Left' | 'LEFT':
        print('Команда ' + cmd.lower())
    case _: # wildcard
        print('другое')
```
