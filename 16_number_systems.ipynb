```python
--1
# перевести двоичное число в десятичную систему счисления
def to_decimai_numbers(number):
    indexes = range(len(str(number)) - 1, -1, -1)
    number = [int(i) for i in str(number)]
    res = 0
    for i in range(len(number)):
        res += number[i] * (2 ** indexes[i])
    return res

number = 111111

print(to_decimai_numbers(number))
```

```python
--3
# перевести шестнадцатеричное число
# в десятичную систему счисления
def to_decimai_numbers(number: str):
    """
    задаем буквы, обозначающие
    в шестнадцатеричной системе числа от 10 до 15
    """
    alphas = ['A', 'B', 'C', 'D', 'E', 'F']
    """
    задаем номера индексов числа, которое необходимо
    перевести в десятичную систему, в обратном порядке
    """
    indexes = range(len(str(number)) - 1, -1, -1)

    res = 0
    for i in range(len(number)):
        if number[i] in alphas:
            res += (10 + alphas.index(number[i])) * (16 ** indexes[i])
        else:
            res += int(number[i]) * (16 ** indexes[i])

    return res

number = '1AF2'

print(to_decimai_numbers(number))
```

```python
--5
# перевести число из десятичной системы в заданную
def from_decimal_system(number: int, system: int):
    res = []
    while number > system - 1:
        res.append(str(number % system))
        number = number // system
    res.append(str(number))

    """
    отдельная проверка для перевода
    в системы с основаниями от 11 до 16
    """
    alphas = ['10', '11', '12', '13', '14', '15']
    for i in range(len(res)):
        if res[i] in alphas:
            if res[i] == '10':
                res[i] = 'A'
            elif res[i] == '11':
                res[i] = 'B'
            elif res[i] == '12':
                res[i] = 'C'
            elif res[i] == '13':
                res[i] = 'D'
            elif res[i] == '14':
                res[i] = 'E'
            else:
                res[i] = 'D'

    return ''.join(res)[::-1]


number, system = 1000, 16

print(from_decimal_system(number, system))
```

```python
--6
# перевести натуральное число в десятичной системе счисления
# в двоичную, восьмеричную и шестнадцатеричную системы счисления
def to_binary_octal_hex(number: int):
    binary_number = bin(number)[2:]
    octal_number = oct(number)[2:]
    hex_number = hex(number)[2:].upper()
    print(binary_number, octal_number, hex_number, sep = '\n')

number = 128

to_binary_octal_hex(number)
```
