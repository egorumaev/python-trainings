```python
--15
number = input()
counter = 0

for index, char in enumerate(number):
    if len(number) < 16:
        print('НЕТ')
        break
    elif index == 0 and char == '+':
        counter += 1
    elif index == 1 and int(char) == 7:
        counter += 1
    elif index == 2 and char == '(':
        counter += 1
    elif index == 6 and char == ')':
        counter += 1
    elif index in [10, 13] and char == '-':
        counter += 1
    elif index in [
        3, 4, 5, 7, 8, \
        9, 11, 12, 14, 15] \
        and char.isdigit() == True:
        counter += 1

print('ДА' if counter == 16 else 'НЕТ')
```

```python
--20
# решето Эратосфена
n = int(input())

# создаем список значений от 0 до n включительно
primes = [i for i in range(n + 1)]
#primes = list(range(n + 1)))

# второй элемент списка - единица,
# которая не является простым числом;
# ставим заглушку
primes[1] = 0

# поиск простых чисел начинаем с числа два
# (это 3-й элемент созданного списка)
i = 2

while i <= n:
    # проверяем: если значение в текущей ячейке
    # до этого не было обнулено,
    # значит в этой ячейке содержится простое число ...
    if primes[i] != 0 and len(primes) > 1:
        #...первое кратное ему будет в два раза больше
        j = i + i
        while j <= n:
            #и это число составное,
            #поэтому заменяем его нулем
            primes[j] = 0
            #переходим к следующему числу,
            #которое кратно i (оно на i больше)
            j = j + i
    i += 1

# избавляемся от всех нулей в списке
primes = [i for i in primes if i != 0]

# проверяем, что согласно условия n
# не входит в итоговый список простых чисел
if n in primes:
    del primes[primes.index(n)]

print(*primes)
```
